```python

# Buat list untuk menampung nilai 
list_nama = []
list_harga = []
list_penjual = []
list_kota_toko = []
list_terjual = []
list_rating =[]

# HASILNYA AKAN ME-RELOAD chrome link itu dari halaman 1-10.

# Looping halaman 1-10
for halaman in range(1, 11):
    # sediakan url
    url = f'https://www.tokopedia.com/search?navsource=&page={halaman}&q=seblak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st='

    # akses website
    driver.get(url)

    # proses loading
    time.sleep(3)       # Tambahkan sedikit waktu untuk makesure halaman dimuat.

    # extract html
    html = driver.page_source

    # Parsing html
    soup = BeautifulSoup(html, 'html.parser')


    # Mengumpulkan box/container items
    items = soup.find_all('div', {'class':'css-1asz3by'})

    # Akses nilai tiap kotak (nama, harga, penjual, kota toko, banyaknya terjual, dan rating)
    for item in items:

        # Mengambil nama produk
        try:
            nama_produk = item.find('div', {'class':'css-3um8ox'}).get_text()
            list_nama.append(nama_produk)
        except:
            list_nama.append(None)

        # Mengambil harga produk 
        try:
            harga_produk = item.find('div', {'class':'css-h66vau'}).get_text()
            list_harga.append(harga_produk)
        except:
            list_harga.append(None)
        

        # Mengambil kota toko 
        try:
            kota_toko = item.find('span', {'class':'prd_link-shop-loc css-1kdc32b flip'}).text if item.find('span', {'class':'prd_link-shop-loc css-1kdc32b flip'}) else None
            list_kota_toko.append(kota_toko)
        except:
            list_kota_toko.append(None)

        # Mengambil penjual 
        try:
            penjual = item.find('span', {'class':'prd_link-shop-name css-1kdc32b flip'}).text if item.find('span', {'class':'prd_link-shop-name css-1kdc32b flip'}) else None
            list_penjual.append(penjual)
        except:
            list_penjual.append(None)


        # Mengambil banyaknya terjual 
        try:
            terjual = item.find('span', {'class':'css-1sgek4h'}).get_text().split(' ')[0]
            list_terjual.append(terjual)
        except:
            list_terjual.append(None)

        # Mengambil rating 
        try:
            rating = item.find('span', {'class':'css-t70v7i'}).get_text()
            if rating:
                list_rating.append(rating)
            else:
                rating = None
        except:
            list_rating.append(None)

```

```python
# Gabungkan list menjadi dataframe
data = pd.DataFrame({
    'Nama Produk': list_nama,
    'Harga Produk' : list_harga,
    'Kota Toko' : list_kota_toko,
    'Penjual' : list_penjual,
    'Banyaknya Terjual' : list_terjual,
    'Rating' : list_rating})

# Panggil datanya
data
```

```python
# Menyimpan DataFrame ke CSV
data.to_csv('data_tokopedia_seblak.csv', index=False)
print("Data berhasil disimpan ke file data_tokopedia_seblak.csv")
```

### 2. Data Cleaning

A. Handle Missing Values

Akan mengganti missing values dengan nilai yang tepat (seperti 0 untuk banyaknya terjual dan rating)

Catatan: Pada kolom "Banyaknya Terjual", jumlah produk yang terjual dituliskan sebagai Terjual 750+, Terjual 1 rb. Ambil angkanya saja. Misal Terjual 750+ -> 750 dan Terjual 1 rb -> 1000.


```python
# Mengganti missing values pada kolom "Banyaknya Terjual" dan "Rating" dengan 0
data['Banyaknya Terjual'] = data['Banyaknya Terjual'].fillna(0)
data['Rating'] = data['Rating'].fillna(0)

# Mengganti missing values pada kolom "Kota Toko" dan "Penjual" dengan 'Unknown'
data['Kota Toko'] = data['Kota Toko'].fillna('Unknown')
data['Penjual'] = data['Penjual'].fillna('Unknown')

```

B. Ubah Bentuk Data Menjadi Konsisten dan Menyesuaikan Tipe Data

Mengubah tipe data pada kolom "Harga Produk" dan "Banyaknya Terjual" menjadi numerik.

```python
# Menghilangkan simbol-simbol non-numerik dan mengubah tipe data menjadi integer

# Mengubah kolom "Harga Produk" menjadi string, membersihkan simbol, lalu mengonversi ke integer
data['Harga Produk'] = data['Harga Produk'].astype(str)
data['Harga Produk'] = data['Harga Produk'].str.replace('Rp', '', regex=False).str.replace('.', '', regex=False).astype(int)

# Mengubah kolom "Banyaknya Terjual" menjadi string, membersihkan simbol, lalu mengonversi ke integer
data['Banyaknya Terjual'] = data['Banyaknya Terjual'].astype(str)
data['Banyaknya Terjual'] = data['Banyaknya Terjual'].str.replace(r'\+', '', regex=True).str.replace(' rb', '000', regex=False)

# Mengatasi nilai kosong atau tidak valid dan konversi ke integer
def convert_to_int(value):
    try:
        return int(value)
    except ValueError:
        return 0

data['Banyaknya Terjual'] = data['Banyaknya Terjual'].apply(convert_to_int)
```

**Penjelasan:**
- Menggunakan str.replace untuk menghapus karakter + dan mengganti rb dengan 000.
- Menambahkan fungsi convert_to_int untuk menangani nilai yang tidak valid dan mengonversinya ke 0.

Dengan ini, semua nilai dalam kolom "Banyaknya Terjual" yang tidak bisa langsung dikonversi akan diubah menjadi 0, sehingga mencegah error saat konversi tipe data.

#### Perhitungan Statistik Deskriptif

```python
# Import Libraries
# scipy.stat utnuk mneghitung nilai skewness dan kurtosis dari data yang ada

import pandas as pd
from scipy.stats import skew, kurtosis

# Membaca data dari file CSV yang telah dibersihkan
data = pd.read_csv('data_tokped_seblak_clean.csv')
```

```python
# Menghitung statistik deskriptif (rata-rata, median, standar deviasi, skewness, dan kurtosis) 

# Dihitung dari kolom harga, banyaknya produk terjual, dan rating. 

columns = ['Harga Produk', 'Banyaknya Terjual', 'Rating']
stats = {}

for col in columns:
    stats[col] = {
        'Mean': data[col].mean(),
        'Median': data[col].median(),
        'Standard Deviation': data[col].std(),
        'Skewness': skew(data[col]),
        'Kurtosis': kurtosis(data[col])
    }

# Membuat DataFrame dari hasil statistik
stats_data = pd.DataFrame(stats)
print(stats_data)
```

### Potensi Pendapatan

Untuk menghitung potensi pendapatan, saya menggunakan data harga produk dan banyaknya terjual. Pendapatan per produk dihitung dengan mengalikan harga produk dengan banyaknya terjual. Dengan asumsi data terdistribusi normal, kita dapat menggunakan confidence interval 95% untuk mendapatkan estimasi pendapatan minimum dan maksimum.

```python
# Menghitung potensi pendapatan 

import pandas as pd
from scipy.stats import norm

# Membaca data dari file CSV yang telah dibersihkan
data = pd.read_csv('data_tokped_seblak_clean.csv')


# Menghitung pendapatan per produk
data['Pendapatan'] = data['Harga Produk'] * data['Banyaknya Terjual']


# Menghitung rata-rata dan standar deviasi pendapatan
mean_pendapatan = data['Pendapatan'].mean()
std_pendapatan = data['Pendapatan'].std()

# Menghitung jumlah sampel
n = len(data)
```

```python
# Menghitung nilai z untuk confidence level 95% (z-1.96)
z = 1.96

# Menghitung margin of error
margin_of_error = z * (std_pendapatan / (n ** 0.5))

# Menghitung confidence interval
ci_lower = mean_pendapatan - margin_of_error
ci_upper = mean_pendapatan + margin_of_error

ci_lower, ci_upper
```

Setelah menghitung interval kepercayaan untuk pendapatan bulanan dari penjualan produk seblak dengan confidence level 95%, kita memperoleh hasil berikut:

- Lower Value (Batas Bawah): 2,562,251.65 rupiah
- Upper Value (Batas Atas): 3,358,106.52 rupiah

Dengan confidence level 95%, kita dapat menyatakan bahwa pendapatan bulanan dari penjualan produk seblak di Tokopedia diperkirakan berada dalam rentang antara 2,562,251.65 rupiah hingga 3,358,106.52 rupiah. Ini berarti, dengan tingkat keyakinan 95%, kita yakin bahwa pendapatan aktual bulanan akan jatuh dalam rentang tersebut.

Rentang ini menunjukkan bahwa ada variasi yang cukup besar dalam pendapatan bulanan, yang bisa jadi disebabkan oleh faktor-faktor seperti variasi harga produk, volume penjualan yang berbeda, dan adanya produk dengan penjualan yang sangat tinggi (outlier).


Dengan potensi pendapatan maksimum yang mencapai lebih dari 3,3 juta rupiah, penjualan seblak di Tokopedia menunjukkan peluang bisnis yang cukup menjanjikan, khususnya jika strategi penjualan dapat meningkatkan volume penjualan atau harga produk rata-rata.

Untuk menentukan apakah ada perbedaan harga barang di Jabodetabek dan di luar Jabodetabek, perlu menggunakan uji hipotesis. 

Langkah-langkah yang akan kita lakukan meliputi:

- Menuliskan Hipotesis Null dan Alternatif
- Memilih Jenis Uji Hipotesis
- Menghitung Statistik Uji dan p-value
- Menarik Kesimpulan Berdasarkan Hasil Uji

1. Menuliskan Hipotesis Null dan Alternatif

    *Hipotesis Null (H0):* Tidak ada perbedaan signifikan antara harga barang di Jabodetabek dan di luar Jabodetabek. 

            H0: H Jabodetabek = H Luar Jabodetabek

    *Hipotesis Alternatif (Ha):* Ada perbedaan signifikan antara harga barang di Jabodetabek dan di luar Jabodetabek. 

            Ha: H Jabodetabek ≠ H Luar Jabodetabek

​
2. Memilih Jenis Uji Hipotesis

    Kami akan menggunakan uji t dua sampel (two-sample t-test) untuk menguji perbedaan rata-rata antara dua kelompok yang tidak berpasangan (independent samples).

3. Menghitung Statistik Uji dan p-value

    Langkah ini melibatkan pembagian data menjadi dua kelompok (Jabodetabek dan luar Jabodetabek), menghitung rata-rata dan standar deviasi masing-masing kelompok, dan kemudian melakukan uji t dua sampel.

```python
# Import library
from scipy.stats import ttest_ind

# Memisahkan data berdasarkan lokasi
jabodetabek = data[data['Kota Toko'].str.contains('Jakarta|Bogor|Depok|Tangerang|Bekasi')]
non_jabodetabek = data[~data['Kota Toko'].str.contains('Jakarta|Bogor|Depok|Tangerang|Bekasi')]

# Melakukan uji-t dua sampel independen
t_stat, p_value = ttest_ind(jabodetabek['Harga Produk'], non_jabodetabek['Harga Produk'], equal_var=False)

# Menampilkan data
t_stat, p_value
```

- t-statistic: -0.02323380370481917
- p-value: 0.9814774645764894

**Analisis**
1. Nilai t-statistic:

    t-statistic yang sangat mendekati 0 (-0.023) menunjukkan bahwa perbedaan rata-rata harga antara Jabodetabek dan luar Jabodetabek sangat kecil.

2. Nilai p-value:

    p-value sebesar 0.981 sangat tinggi dan jauh di atas ambang batas umum 0.05. Ini berarti, probabilitas untuk mendapatkan hasil seperti ini atau lebih ekstrem jika hipotesis null benar sangat tinggi.

**Kesimpulan**

Karena p-value (0.981) lebih besar dari tingkat signifikansi 0.05, jadi kami gagal menolak hipotesis null. Ini berarti kami tidak memiliki bukti yang cukup untuk menyatakan bahwa terdapat perbedaan yang signifikan antara harga barang di Jabodetabek dan di luar Jabodetabek.

**Insight**
1. Harga Barang: Dari hasil analisis statistik, kami dapat menyimpulkan bahwa secara umum, harga barang di Jabodetabek dan di luar Jabodetabek tidak menunjukkan perbedaan yang signifikan.
2. Strategi Penetapan Harga: Informasi ini berguna bagi kami (penjual) dan distributor dalam menetapkan strategi harga. Karena tidak ada perbedaan signifikan dalam harga, strategi penetapan harga yang konsisten di berbagai lokasi dapat dipertimbangkan.
3. Faktor Lain: Meskipun tidak ada perbedaan yang signifikan dalam harga, kami mungkin perlu mempertimbangkan faktor lain seperti biaya pengiriman, ketersediaan produk, dan preferensi konsumen di lokasi yang berbeda untuk mengoptimalkan penjualan.

Selanjutnya, untuk mengetahui apakah orang lebih cenderung menyukai produk yang harganya murah, kami menggunakan uji korelasi antara harga produk dan rating produk.

Langkah Perhitungan:

1. Menghitung nilai korelasi dan p-value:
Kami akan menggunakan korelasi Pearson jika data terdistribusi normal, atau Spearman jika data tidak terdistribusi normal.
2. Analisis hasil:
    - Korelasi positif menunjukkan bahwa ketika harga naik, rating juga naik.
    - Korelasi negatif menunjukkan bahwa ketika harga naik, rating turun.
    - p-value yang kecil (biasanya < 0.05) menunjukkan bahwa hasil korelasi signifikan.

```python
# Import libraries
import pandas as pd
from scipy.stats import skew, kurtosis, pearsonr, spearmanr

# Membaca data dari file CSV yang telah dibersihkan
data = pd.read_csv('data_tokped_seblak_clean.csv')

# Menghitung skewness dan kurtosis
harga_skewness = skew(data['Harga Produk'])
harga_kurtosis = kurtosis(data['Harga Produk'])

rating_skewness = skew(data['Rating'])
rating_kurtosis = kurtosis(data['Rating'])

# hasil perhitungan
harga_skewness, harga_kurtosis, rating_skewness, rating_kurtosis
```

Dengan mempertimbangkan hasil dari uji skewness dan kurtosis yang menunjukkan bahwa data harga dan rating tidak terdistribusi normal (harga sangat skewed positif dan kurtosis tinggi, rating sangat skewed negatif dan kurtosis tinggi), kami akan menggunakan uji korelasi Spearman. Spearman cocok untuk data yang tidak terdistribusi normal.

```python
# Menghitung korelasi Spearman antara harga produk dan rating
corr, p_value = spearmanr(data['Harga Produk'], data['Rating'])

# Tampilkan hasil
corr, p_value
```

Sesuai dengan hasil perhitungan: 

- Nilai Korelasi (corr): 0.134
- p-value: 0.011

**Kesimpulan**

Dari hasil korelasi Spearman, dapat disimpulkan bahwa ada kecenderungan konsumen memberikan rating yang sedikit lebih tinggi untuk produk dengan harga yang lebih tinggi, meskipun hubungannya lemah. Korelasi positif yang signifikan menunjukkan bahwa ketika harga produk seblak meningkat, rating yang diberikan oleh konsumen cenderung sedikit meningkat juga.

Ini memberikan insight bahwa harga produk mungkin mempengaruhi persepsi konsumen terhadap kualitas produk. Konsumen mungkin mengasosiasikan harga yang lebih tinggi dengan kualitas yang lebih baik, sehingga mereka memberikan rating yang lebih tinggi.
